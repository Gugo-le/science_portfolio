```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.show()
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.title('students') 
plt.plot([1,2,3,4],[2,4,8,6]) 
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10]) # 라 인 plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student']) # 범 례 plt.show()
```
